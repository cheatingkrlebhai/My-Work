```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Initialize the Chrome driver
driver = webdriver.Edge()

# Login process
driver.get('https://www.linkedin.com/login/')
time.sleep(3)

# Enter the username
username_field = driver.find_element(By.ID, "username")
username_field.send_keys("USERNAME")

# Enter the password
password_field = driver.find_element(By.ID, "password")
password_field.send_keys("PASSWORD")

login_button = driver.find_element(By.XPATH, "//*[@id='organic-div']/form/div[3]/button")
login_button.click()

# Wait for the homepage to load
time.sleep(10)

# The message to send
message = """
Greetings,

I am running a student-led business consulting group comprised of students from elite colleges like IIMs, IITs, SRCC, SSCBS etc. We were intrigued by your company and were wondering to collaborate with you for a Live Business Consulting Project
"""

# Loop through pages 22 to 30
for page in range(2, 21):
    # Construct the URL for each page
    url = f'https://www.linkedin.com/search/results/people/?origin=FACETED_SEARCH&page={page}&schoolFilter=%5B%2212070%22%2C%22215708%22%2C%225176%22%2C%22157272%22%2C%22157273%22%2C%22157276%22%5D&sid=4bq&titleFreeText=Founder'
    driver.get(url)
    
    # Wait for the page to load
    time.sleep(5)

    # Find all "Connect" buttons
    all_buttons = driver.find_elements(By.TAG_NAME, "button")
    connect_buttons = [btn for btn in all_buttons if btn.text == "Connect"]

    for btn in connect_buttons:
        driver.execute_script("arguments[0].click();", btn)
        time.sleep(2)
        
        try:
            # Click the "Add a note" option
            add_note_button = driver.find_element(By.XPATH, "//button[@aria-label='Add a note']")
            driver.execute_script("arguments[0].click();", add_note_button)
            time.sleep(2)
            
            # Enter the message
            message_box = driver.find_element(By.XPATH, "//textarea[@name='message']")
            message_box.send_keys(message)
            time.sleep(2)

            # Click the "Send" button
            send_button = driver.find_element(By.XPATH, "/html/body/div[3]/div/div/div[3]/button[2]")
            driver.execute_script("arguments[0].click();", send_button)
            time.sleep(2)
        except Exception as e:
            print(f"An error occurred on page {page}:", e)
        
        time.sleep(2)
    
    print(f"Completed page {page}")
    time.sleep(5)  # Wait before moving to the next page

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Initialize the Chrome driver
driver = webdriver.Edge()

# Login process
driver.get('https://www.linkedin.com/login/')
time.sleep(3)

# Enter the username
username_field = driver.find_element(By.ID, "username")
username_field.send_keys("USERNAME")

# Enter the password
password_field = driver.find_element(By.ID, "password")
password_field.send_keys("PASSWORD")

login_button = driver.find_element(By.XPATH, "//*[@id='organic-div']/form/div[3]/button")
login_button.click()

# Wait for the homepage to load
time.sleep(10)

# Loop through pages 22 to 30
for page in range(22, 31):
    # Construct the URL for each page
    url = f'https://www.linkedin.com/search/results/people/?geoUrn=%5B%22102454443%22%5D&keywords=Co%20founder&network=%5B%22S%22%5D&origin=FACETED_SEARCH&page={page}&sid=%40_t'
    driver.get(url)
    
    # Wait for the page to load
    time.sleep(5)

    # Find all "Connect" buttons
    all_buttons = driver.find_elements(By.TAG_NAME, "button")
    connect_buttons = [btn for btn in all_buttons if btn.text == "Connect"]

    for btn in connect_buttons:
        try:
            driver.execute_script("arguments[0].click();", btn)
            time.sleep(2)
            
            # Check if there's a "Send" button directly available
            send_button = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.XPATH, "/html/body/div[3]/div/div/div[3]/button[2]"))
            )
            driver.execute_script("arguments[0].click();", send_button)
            
            print("Connection request sent successfully")
            time.sleep(2)
        except Exception as e:
            print(f"An error occurred:", e)
        
        time.sleep(2)
    
    print(f"Completed page {page}")
    time.sleep(5)  # Wait before moving to the next page

# Close the browser
driver.quit()
```
